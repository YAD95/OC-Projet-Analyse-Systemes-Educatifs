```python
def calcul_proportion_annee (annee):
    total_valeur = int(data_annees_filtre.shape[0])
    total_null = data_annees_filtre[annee].isna().sum()                   
    annee_proportion = (100- (total_null / total_valeur) *100).sort_values(ascending=False)
    return annee_proportion
annee_data = [c for c in data_annees_filtre.columns if c.isdigit()]
print(calcul_proportion_annee(annee_data))
# pourcentage par années des valeur rémplis 
```

Les années présentant le plus de valeurs renseignées ont été identifiées afin de conserver un minimum de qualité des données.

## Étape 1 - Identifiez les indicateurs redondants d’un point de vue statistique

### 1) Calculez une matrice de corrélation entre l’ensemble des indicateurs en utilisant la méthode de Pearson puis de Spearman

```python
corr_pearson = df_pays_indicateurs.corr(method="pearson")
corr_pearson
```

```python
corr_spearman  = df_pays_indicateurs.corr(method="spearman")
corr_spearman
```

Les coefficients de corrélation de Pearson et de Spearman permettent d’identifier des relations statistiques entre indicateurs, mais ils ne permettent pas de montrer une relation de causalité. Pearson est adapté à la détection de relations linéaires tandis que Spearman permet de détecter des relations monotones, y compris non linéaires. En revanche, ces coefficients ne capturent pas les relations complexes ou non monotones et peuvent être influencés par la présence de valeurs aberrantes.

### 2) Visualisez les résultats avec une heatmap

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

mask = np.triu(np.ones_like(corr_pearson, dtype=bool))

plt.figure(figsize=(12, 10))
sns.heatmap(
    corr_pearson,
    mask=mask,
    vmin=-1,
    vmax=1,
    center=0,
    cmap="coolwarm"
)
plt.title("Matrice de corrélation de Pearson")
plt.show()
```

```python
mask = np.triu(np.ones_like(corr_spearman, dtype=bool))

plt.figure(figsize=(12, 10))
sns.heatmap(
    corr_spearman,
    mask=mask,
    vmin=-1,
    vmax=1,
    center=0,
    cmap="coolwarm"
)
plt.title("Matrice de corrélation de Spearman")
plt.show()
```

La comparaison entre les coefficients de Pearson et de Spearman montre que certaines relations présentent une corrélation plus élevée selon Spearman que selon Pearson. Cela suggère l’existence de relations monotones mais non strictement linéaires, comme par exemple entre la population totale et le PIB, où les pays les plus peuplés ont tendance à avoir un PIB plus élevé sans que cette relation soit proportionnelle

- le PIB total, représentant la taille globale du marché ;
- la population totale, indiquant le potentiel démographique du pays.

Les indicateurs liés au nombre d’élèves ont été pondérés plus fortement, car ils constituent le cœur du modèle économique de l’academy. Les pondérations retenues sont les suivantes :
- élèves (secondaire + université) : coefficient 2

- population totale : coefficient  1

Les indicateurs étant exprimés dans des ordres de grandeur très différents, une normalisation de type Min-Max a été appliquée afin de rendre les variables comparables. Un score global a ensuite été calculé comme une combinaison linéaire pondérée de ces indicateurs.
